```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('healthcare-dataset-stroke-data.csv')
dataset.head(10)
```

```python
dataset['bmi'] = dataset['bmi'].fillna(0)

ave = dataset['bmi'].mean()
dataset['bmi'] = dataset['bmi'].replace(0, ave) #Replacing zeros with average BMI
```

```python
dataset.drop(columns=['id'],inplace=True)
```

```python
#Converting Gender to categorical
dataset.head(10)
dataset['gender'] = dataset['gender'].astype('category')
dataset['gender'] = dataset['gender'].cat.codes
dataset.head(10)
```

```python
#Converting marital status to categorical
dataset['ever_married'] = dataset['ever_married'].astype('category')
dataset['ever_married'] = dataset['ever_married'].cat.codes
```

```python
#Converting work type, residence type and smoking status to categorical
dataset['work_type'] = dataset['work_type'].astype('category')
dataset['work_type'] = dataset['work_type'].cat.codes
dataset['Residence_type'] = dataset['Residence_type'].astype('category')
dataset['Residence_type'] = dataset['Residence_type'].cat.codes
dataset['smoking_status'] = dataset['smoking_status'].astype('category')
dataset['smoking_status'] = dataset['smoking_status'].cat.codes
```

```python
dataset.head(10)
```

```python
#Seperating Features and dependant variable
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=0)
```

```python
#Feature scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test) 
```

```python
#Simple Logistic Regression Model
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state=0)
classifier.fit(X_train,y_train)

classifier.predict(sc.transform([[1,75.0,0,1,0,2,1,5000,36,1]]))
y_pred = classifier.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))


from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test,y_pred)
print(cm)

accuracy_score(y_test,y_pred)
```

```python
#Naive Bayes Model
from sklearn.naive_bayes import GaussianNB
classifier1 = GaussianNB()
classifier1.fit(X_train,y_train)

classifier1.predict(sc.transform([[1,75.0,0,1,0,2,1,5000,36,1]]))

y_pred = classifier1.predict(X_test)
print(np.concatenate((y_pred.reshape(len(y_pred),1),y_test.reshape(len(y_test),1)),1))

cm = confusion_matrix(y_test,y_pred)
print(cm)

accuracy_score(y_test,y_pred)
```
